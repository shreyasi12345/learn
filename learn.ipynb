```python
import pandas as pd
import numpy as np

# Creating a messy DataFrame
data = {
    "ID": [101, 102, 103, 104, 105, np.nan, 107, 108, 109, 110],
    "Name": ["Alice", "Bob", "Charlie", "David", np.nan, "Frank", "Eve", "Grace", "Hannah", "Isaac"],
    "Age": [25, np.nan, 30, 22, 28, 35, 29, np.nan, 40, 31],
    "Salary": [50000, 60000, np.nan, 52000, 48000, 55000, 57000, 62000, np.nan, 59000],
    "Department": ["HR", "Finance", "IT", "IT", "HR", "Finance", "HR", np.nan, "IT", "Finance"]
}

df = pd.DataFrame(data)

print(df)
```

```python
#missing values in each column
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
```

```python
df.dropna()
```

```python
threshold= len(df)*0.5
df_cleaned = df.dropna(axis=1, thresh=threshold)
df_cleaned
```

```python
df_cleaned["Salary"].fillna(df_cleaned["Salary"].median(),inplace=True)
df_cleaned
```

```python
df.duplicated().sum()
```

```python
df_cleaned=df_cleaned.drop_duplicates(subset="ID")
df_cleaned
```
