```python
import pandas as pd
import numpy as np

# Creating a DataFrame with many missing values
data3 = {
    "Customer_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Alice", "Bob", "Charlie", "David", None, "Frank", "Grace", "Helen", None, "Jack"],
    "Age": [25, np.nan, 30, 45, np.nan, 40, np.nan, 35, np.nan, 50],
    "Gender": ["F", "M", None, "M", "F", None, "F", "F", "M", None],
    "Salary": [50000, 60000, None, 70000, 55000, None, 45000, None, 65000, 52000],
    "City": ["New York", "Los Angeles", None, "Chicago", "Houston", None, "Miami", "Boston", "San Francisco", None]
}

df3 = pd.DataFrame(data3)
df3
```

```python
df3.isnull().sum()
```

```python
df3['Salary'] = df3['Salary'].fillna(df3['Salary'].mean())
df3
```

```python
df3['Age'] = df3['Age'].fillna(df3['Age'].mean())
df3
```

```python
df3['Gender'] = df3['Gender'].fillna(df3['Gender'].mode()[0])
df3
```

```python
df3['City'] = df3['City'].fillna(df3['City'].mode()[0])
df3
```

```python
df3.drop_duplicates()
```

```python
df3_dropped=df3.dropna()
df3_dropped
```

```python
df3_dropped1=df3.dropna(subset=["Name","Age"])
df3_dropped1
```

```python
threshold=0.4*len(df3)
df3_dropped2=df3.dropna(axis=1,thresh=threshold)
```

```python
x = [[1, 2], [3, 4]]  # Nested list
y = x[:]  # Shallow copy using slicing

# Modify the nested list in y
y[0][0] = 99

print(x)  # Original list
print(y)  # Copied list
```

```python
x = [[1], 2, 3]  # Mixed list (nested list + integers)
y = x[:]  # Shallow copy
y[0][0] = 99  # Adding a new element to y

print(x)  # Original list
print(y)  # Copied list
```
