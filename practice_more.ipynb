```python

data = {
    "ID": [101, 102, 103, 104, 105, np.nan, 107, 108, 109, 110],
    "Name": ["Alice", "Bob", "Charlie", "David", np.nan, "Frank", "Eve", "Grace", "Hannah", "Isaac"],
    "Age": [25, np.nan, 30, 22, 28, 35, 29, np.nan, 40, 31],
    "Salary": [50000, 60000, np.nan, 52000, 48000, 55000, 57000, 62000, np.nan, 59000],
    "Department": ["HR", "Finance", "IT", "IT", "HR", "Finance", "HR", np.nan, "IT", "Finance"]
}

df = pd.DataFrame(data)

df
```

```python
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df
```

```python
df["Salary"]=df["Salary"].fillna(df["Salary"].mean())
df
```

```python
df["Department"]=df["Department"].fillna(df["Department"].mode()[0])
df
```

```python
df.drop_duplicates()
```

```python
df.dropna()
```

```python
avg_salary=df.groupby("Department")["Salary"].mean()
avg_salary
```

```python
count_pep=df.groupby("Department").value_counts()
count_pep
```

```python
df.sort_values(["Salary"])
```

```python
df[df["Salary"]>50000]
```
