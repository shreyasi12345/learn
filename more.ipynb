```python
import pandas as pd
import numpy as np

# Creating a new DataFrame with some data
data2 = {
    "Product_ID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010],
    "Product_Name": ["Apple", "Banana", "Carrot", "Date", "Eggplant", "Fig", "Grape", "Honeydew", "Indian Fig", "Jalapeno"],
    "Category": ["Fruit", "Fruit", "Vegetable", "Fruit", "Vegetable", "Fruit", "Fruit", "Fruit", "Fruit", "Vegetable"],
    "Price": [1.2, 0.5, 0.7, 2.5, 1.8, 2.0, 3.0, 4.0, 2.5, 1.5],
    "Stock": [120, 150, 80, 60, 30, 100, 90, 20, 50, 110],
    "Discount": [0.1, 0.05, 0.15, 0.2, 0.25, 0.1, 0.15, 0.05, 0.1, 0.2]
}

df2 = pd.DataFrame(data2)
df2
```
